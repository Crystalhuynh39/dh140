```python
# for loop
k = 0 # we want to start at the first row
hit_song = 0
popular_artist = 0
hit_song_and_popular_artist = 0
for i in music_df.iterrows():
    artist_hot = (music_df['artist.hotttnesss'])[k] # this is the value for an artist's hotness or popularity
    song_hot = (music_df['song.hotttnesss'])[k] # this is the value for the artist's song's hotness or popularity
    if artist_hot == song_hot:
        #print((music[k])[6] + " is a popular artist that makes hit songs")
        hit_song_and_popular_artist+=1
    elif artist_hot > song_hot:
        #print((music_df['artist.name'])[k] + " is a popular artist")
        popular_artist+=1
    elif artist_hot < song_hot:
        #print((music_df['artist.name'])[k] + " makes hit songs")
        hit_song+=1
    k+=1

print(hit_song_and_popular_artist, hit_song, popular_artist) 
```

```python
artist_status_dict = {'artist_status': ['make hit songs and are popular', 'make hit songs', 'are a popular artist'],
            'count': [hit_song_and_popular_artist, hit_song, popular_artist]
    }

artist_status_df = pd.DataFrame(artist_status_dict)

artist_status_bar = artist_status_df.plot(kind='bar',x='artist_status', y='count', title = 'How Many Artists Only Make Hit Songs vs. Are Only Popular vs. Are Both')
```

```python
artist_status_pie = artist_status_df.plot(kind='pie', y='count', labels = artist_status_df['count'], title = 'How Many Artists Only Make Hit Songs vs. How Many Artists Are Only Popular')
artist_status_pie.legend(artist_status_df['artist_status'])
```

We can see that there are a lot of artists that are considered popular but don't necessarily make hit songs. There are less artists that make hit songs but are not considered popular. There are even less artists who fit both criteria in that they are both popular and make hit songs. This makes sense as the music industry is very saturated and those who fit both criteria have the "it" factor that make them stand out among others, and are thus considered superstars.
